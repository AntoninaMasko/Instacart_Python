## Task 6

The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

    If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
    
    If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
# Creating a "mean_spend" column:
ords_prods_merge['mean_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check of the new column:
ords_prods_merge.head(20)
```

```python
# Creating a flag that assigns a spending flag:
ords_prods_merge.loc[ords_prods_merge['mean_spend'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_spend'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Checking the frequency of the new column
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Verify that the flag was assigned correctly:
ords_prods_merge[['user_id', 'spending_flag', 'mean_spend']].head(100)
```

## Task 7

In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

    If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
    
    If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
    
    If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
# Creating a "median_order_frequency" column:
ords_prods_merge['median_order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_previous_order'].transform(np.median)
```

```python
# Check of the new column:
ords_prods_merge.head(20)
```

```python
# Creating a flag that assigns a buying frequency flag:
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order_frequency'] > 10) & (ords_prods_merge['median_order_frequency'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Checking the frequency of the new column
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Creating a subset with null values in the frequency flag column:
freq_nan = ords_prods_merge[ords_prods_merge['frequency_flag'].isnull()]
```

```python
freq_nan.head()
```

The null values (NaN) appear in the "frequency_flag" column due to null values in the "day_since_previous_order". We can see that this is a new customer.

```python
# Creating a subset for customer_id 159838 to see their entire history:
user_id_159838 = ords_prods_merge[ords_prods_merge['user_id'] == '159838']
```

```python
user_id_159838
```

From the above subset, we see that this customer made only one order, therefore it is normal that there is a null value in the "days_since_previous_order" and the median is not calculated. In this situation, I propose to add an additional criterium to flag the frequency.

```python
# Creating additional flag for the NaN values in the "median_order_frequency": 
ords_prods_merge.loc[ords_prods_merge['median_order_frequency'].isnull(), 'frequency_flag'] = 'Non-frequent customer'
```

```python
# Checking again the frequency of the new column
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Verify that the flag was assigned correctly:
ords_prods_merge[['user_id', 'frequency_flag', 'median_order_frequency']].head(100)
```

After additional verification and aggregation, now we have the frequency flag assigned correctly.
