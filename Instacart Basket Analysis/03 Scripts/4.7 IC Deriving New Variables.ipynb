# 03. If-Statements with User-Defined Functions 

```python
df = df_ords_prods[:1000000]
```

```python
# Creatig a user function:
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df.shape
```

```python
# Applying the user function to the dataframe
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Checking the highest price
df['prices'].max()
```

# 05. If-Statements with For-Loops

```python
# Checking the frequency of the "orders_day_of_week"
df_ords_prods['orders_day_of_week'].value_counts(dropna=False)
```

```python
# Creating a for-loop to assign the busy day values:
result = []

for day in df_ords_prods['orders_day_of_week']:
    if day == 0:
        result.append('Busiest day')
    elif day == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

```python
# Checking the result list:
result
```

```python
# Adding the "result" list to the dataframe:
df_ords_prods['busiest_day'] = result
```

```python
# Checking the frequency of the "busiest_day" column:
df_ords_prods['busiest_day'].value_counts(dropna = False)
```

The results are as expected and in line with the frequency check ran earlier.

Suppose your clients have changed their minds about the labels you created in your busiest_day column. Now, they want Busiest day to become Busiest days (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, theyd also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# Creating a for-loop to assign the 2 busies and slowest days values:
result_2_days = []

for day in df_ords_prods['orders_day_of_week']:
    if day == 0:
        result_2_days.append('Busiest day')
    elif day == 1:
        result_2_days.append('Busiest day')
    elif day == 4:
        result_2_days.append('Least busy')
    elif day == 3:
        result_2_days.append('Least busy')
    else:
        result_2_days.append('Regularly busy')
```

```python
# Adding the newly created list to the dataframe:
df_ords_prods['busiest_days'] = result_2_days
```

## Task 4

When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled Most orders, Average orders, and Fewest orders. Create a new column containing these labels called busiest_period_of_day.

```python
# Checking the frequency of the "order_hour_of_day"
df_ords_prods['order_hour_of_day'].value_counts(dropna=False)
```

After analysing the above frequency, I have decided to create the following time periods (including both hours): 
    10 - 15: most orders; 
    0 - 6: fewest orders; 
    7 - 9 and 16 - 23: average orders

```python
# Creating a for-loop to assign the busies and slowest hours of the day values:
result_hours = []

for hour in df_ords_prods['order_hour_of_day']:
    if hour >= 10 and hour <= 15:
        result_hours.append('Most orders')
    elif hour >= 0 and hour <= 5:
        result_hours.append('Fewest orders')
    else:
        result_hours.append('Average orders')
```

```python
# Checking the result of the newly created list:
result_hours
```

```python
# Adding the newly created list to the dataframe:
df_ords_prods['busiest_period_of_day'] = result_hours
```
